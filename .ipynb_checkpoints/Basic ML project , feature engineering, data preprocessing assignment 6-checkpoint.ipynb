### data cleaning

```python
data[data.isnull().any(axis=1)]
```

Many countries have missing values for the prev_medals column as many countries did not participate in the olympics for example Albania, did not participate in 1988 Olympics

```python
# drop the rows with missing values
data = data.dropna()
```

```python
data
```

```python
# split the data : last 2 years of data (2012 and 2016) will be the testing data and previous years will be out testing data
train = data[data['year']<2012].copy()
test = data[data['year']>=2012].copy()
```

```python
train.shape
```

```python
test.shape
```

```python
# training the model
from sklearn.linear_model import LinearRegression

reg=LinearRegression()
```

```python
predictors = ['athletes','prev_medals']
target = 'medals'
```

```python
reg.fit(train[predictors], train['medals'])
```

```python
# make predictions
predictions = reg.predict(test[predictors])
```

```python
predictions
```

the predictions are not a whole number and some predictions are in negative
so, we need to rescale our predictions so that they make sense

```python
test['predictions']=predictions
```

```python
test
```

```python
# predictions whichw are negative will be replaced with 0
test.loc[test['predictions']<0,'predictions']=0
```

```python
# the predictions will be rounded off to the nearesr whole number
test['predictions'] = test['predictions'].round()
```

```python
from sklearn.metrics import mean_absolute_error

error = mean_absolute_error(test['medals'], test['predictions'])
```

```python
error
```

```python
data.describe()['medals']
```

our mean absolute error should be below the standard deviation. If it is above the standard devistion, usually something is wrong - either we have used the predictors that do not give any information to predict the targe or something gas gone wrong with the model.

Here, our error is quite below the standard deviation

```python
test[test['team']=='USA']
```

```python
test[test['team']=='IND']
```

```python
# look at errors by country

errors = (test['medals']-test['predictions']).abs()
```

```python
errors
```

```python
error_by_team = errors.groupby(test['team']).mean()
```

```python
error_by_team
```

```python
medals_by_team = test['medals'].groupby(test['team']).mean()
```

```python
error_ratio = error_by_team / medals_by_team
```

```python
error_ratio
```

```python
error_ratio[~pd.isnull(error_ratio)]
```

```python
# clean up the infinite values
import numpy as np
error_ratio = error_ratio[np.isfinite(error_ratio)]
```

```python
error_ratio
```

```python
error_ratio.plot.hist()
```

interpretation : error ratio of 0 to 0.5 means you are within 50% of actual medal count. But in some cases, we have error ratio of 2 and aboove which means our predictions were twice as high or half the actual medals
